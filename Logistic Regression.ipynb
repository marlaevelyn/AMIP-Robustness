# 1. Generate Data

```r
n <- 5000
theta <- 2
intercept <- 2
x_mu <- 1
x_var <- 1

df <- data.frame("x" = rnorm(n, x_mu, x_var))
df$z <- theta*df$x + intercept
df$p <- 1/(1+exp(-df$z))
df$y <- rbinom(1,1,df$p)

log_model <- glm(formula="y ~ x", family="binomial"(link='logit'), data=df)

plot(df$x, df$p) + abline(log_model)
```
